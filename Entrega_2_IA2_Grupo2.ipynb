```python
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
X
```

```python
y
```

```python

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

sc2 = StandardScaler()
y_train = sc2.fit_transform(y_train.reshape(-1,1))
y_test = sc2.transform(y_test.reshape(-1,1))

media_y = y.mean()
desvio_estandar_y = y.std()
```

```python
X_test
```

```python
y_test
```

```python
red = tf.keras.models.Sequential()
red.add(tf.keras.layers.Dense(units=12, activation='relu'))
red.add(tf.keras.layers.Dense(units=6, activation='relu'))
red.add(tf.keras.layers.Dense(units=1, activation='linear'))
red.compile(optimizer = tf.keras.optimizers.SGD(learning_rate=0.3),
            loss = 'mse',
            metrics = ['mse', 'mae'])

# ENTRENAMOS
red.fit(X_train, y_train, batch_size = 32, epochs = 2000)
```

```python
predecido = red.predict(sc.transform([
    [425.783283, 144.743682, 397.851502 ]]
                         )
            )[0][0]

prediccion_ultimo_registro = predecido*desvio_estandar_y + media_y
```

```python
prediccion_ultimo_registro
```

```python
y_pred = red.predict(X_test)

for i in range(len(y_pred)):
    error_absoluto = abs(y_test[i][0] - y_pred[i][0])
    error_relativo = abs(error_absoluto / y_test[i][0])*100

    line_info = f"ERROR RELATIVO % {error_relativo} || ERROR ABSOLUTO : {error_absoluto}"
    print(line_info)
```
